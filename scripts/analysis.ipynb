```python
from pathlib import Path
import sys
import re
import pandas as pd
```

```python
folder = Path("/Users/samiraisin/Devil_semelparity_project/store_results/18.03.2026/")
```

```python
filenames = [f.name for f in folder.iterdir() if f.suffix == ".csv"]
```

```python
len(filenames)
```

```python
m = re.search(r"^INF1_([\d.]+)_INF2_([\d.]+)_sample_(\d+)_statistics.csv$", 'INF1_0.5_INF2_0.0_sample_1_statistics.csv')

if m:
    print(m.groups())
```

```python
df = pd.DataFrame(columns=["INF1", "INF2", "sample", "result"])
```

```python
for filename in filenames:
    m = re.search(r"^INF1_([\d.]+)_INF2_([\d.]+)_sample_(\d+)_statistics.csv$", filename)
    if m:
        INF1, INF2, sample = m.groups()
        result = pd.read_csv(folder / filename)['result'].iloc[-1]
        df = pd.concat([df, pd.DataFrame([{"INF1": float(INF1), "INF2": float(INF2), "sample": int(sample), "result": result}])], ignore_index=True)
```
