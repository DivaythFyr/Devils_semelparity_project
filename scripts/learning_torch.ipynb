```python
# Standard numpy-like indexing and slicing
tensor = torch.ones(4, 4)
print(f"Tensor original: {tensor}")

print(f"First row: {tensor[0]}")
print(f"First column: {tensor[:, 0]}")
print(f"Last column: {tensor[..., -1]}")

print()
tensor[:,1] = 0
print(tensor, "after setting second column to zero")
```

```python
# Joining tensors You can use torch.cat to concatenate a sequence of tensors
# along a given dimension.
# See also torch.stack, another tensor joining operator
# that is subtly different from torch.cat.

t1 = torch.cat([tensor, tensor, tensor], dim=1)
print(t1)
```

```python
# This computes the matrix multiplication between two tensors. y1, y2, y3 will have the same value
# ``tensor.T`` returns the transpose of a tensor
y1 = tensor @ tensor.T
y2 = tensor.matmul(tensor.T)

y3 = torch.rand_like(y1)
torch.matmul(tensor, tensor.T, out=y3)
```

```python
# This computes the element-wise product. z1, z2, z3 will have the same value
z1 = tensor * tensor
z2 = tensor.mul(tensor)

z3 = torch.rand_like(tensor)
torch.mul(tensor, tensor, out=z3)
```

```python
#In-place operations Operations that store the result into the operand
# are called in-place. They are denoted
# by a _ suffix. For example: x.copy_(y), x.t_(), will change x.

print(f"{tensor} \n")
tensor.add_(5)
print(tensor)
```

```python
# Method 1: Direct boolean tensor
mask1 = torch.tensor([True, False, True])

# Method 2: Comparison operations (most common)
tensor = torch.tensor([1, 2, 3, 4, 5])
mask2 = tensor > 3  # [False, False, False, True, True]

# Method 3: Logical operations
mask3 = (tensor > 2) & (tensor < 5)  # AND: [False, False, True, True, False]
mask4 = (tensor < 2) | (tensor > 4)  # OR:  [True, False, False, False, True]
mask5 = ~(tensor == 3)               # NOT: [True, True, False, True, True]
```

```python
# torch.nonzero() returns the indices of
# all non-zero elements in a tensor.
# It's extremely useful for finding where specific conditions
# are met in your data.
tensor_1d = torch.tensor([0, 5, 0, 3, 0, 8])
torch.nonzero(tensor_1d)

# tensor([[1],
#         [3],
#         [5]])
# Meaning: elements at positions 1, 3, and 5 are non-zero
```

```python
tensor_2d = torch.tensor([[0, 1, 0],
                       [2, 0, 3],
                       [0, 0, 4]])
torch.nonzero(tensor_2d)

# tensor([[0, 1],  # Row 0, Column 1 (value: 1)
#         [1, 0],  # Row 1, Column 0 (value: 2)
#         [1, 2],  # Row 1, Column 2 (value: 3)
#         [2, 2]]) # Row 2, Column 2 (value: 4)
```

```python
import torch
import numpy as np
import time

def benchmark_operation(sizes, device='mps'):
    results = {'numpy': [], 'pytorch': []}
    
    for N in sizes:
        # NumPy version
        pos_np = np.random.rand(N, 2).astype(np.float32)
        
        t0 = time.perf_counter()
        for _ in range(10):  # Average over runs
            dx = pos_np[:, 0][:, None] - pos_np[:, 0][None, :]
            dy = pos_np[:, 1][:, None] - pos_np[:, 1][None, :]
            dist = np.sqrt(dx**2 + dy**2)
        numpy_time = (time.perf_counter() - t0) / 10
        results['numpy'].append(numpy_time)
        
        # PyTorch version
        pos_torch = torch.from_numpy(pos_np).to(device)
        
        # Warmup (important for GPU!)
        for _ in range(3):
            dist = torch.cdist(pos_torch, pos_torch)
        
        t0 = time.perf_counter()
        for _ in range(10):
            dist = torch.cdist(pos_torch, pos_torch)
            torch.cuda.synchronize() if device=='cuda' else None
        pytorch_time = (time.perf_counter() - t0) / 10
        results['pytorch'].append(pytorch_time)
        
        speedup = numpy_time / pytorch_time
        print(f"N={N:5d}: NumPy={numpy_time*1000:6.1f}ms, "
              f"PyTorch={pytorch_time*1000:6.1f}ms, "
              f"Speedup={speedup:.2f}x")
    
    return results

# Run benchmark
sizes = [500, 1000, 2000, 3000, 5000, 7000, 10000, 15000, 20000]
results = benchmark_operation(sizes, device='mps')  # or 'cuda'
```
